## ğŸ“¥ å°†æ•™ç¨‹ Notebooks å¤åˆ¶åˆ°æ‚¨çš„ç©ºé—´ï¼ˆå¯é€‰ï¼‰

è¿è¡Œæ­¤å•å…ƒæ ¼ä»¥å°†æ‰€æœ‰æ•™ç¨‹ notebook å¤åˆ¶åˆ°æ‚¨çš„ä¸ªäººå·¥ä½œç©ºé—´ï¼Œä»¥ä¾¿æ‚¨å¯ä»¥ä¿®æ”¹å®ƒä»¬ï¼š

```python
# å°†æ•™ç¨‹ notebook å¤åˆ¶åˆ°æ‚¨çš„å·¥ä½œç©ºé—´
import shutil
from pathlib import Path

shared_notebooks = Path('/shared/notebooks')
my_notebooks = Path.home() / 'work' / 'notebooks'

# è¦å¤åˆ¶çš„ notebook åˆ—è¡¨
tutorials = [
    '01_quickstart.ipynb',
    '02_variant_analysis.ipynb',
    '03_batch_analysis.ipynb',
    '04_visualization.ipynb',
    '05_custom_analysis.ipynb',
]

copied = []
for notebook in tutorials:
    src = shared_notebooks / notebook
    dst = my_notebooks / notebook
    if src.exists() and not dst.exists():
        shutil.copy(src, dst)
        copied.append(notebook)
    elif dst.exists():
        print(f"âš ï¸  {notebook} å·²å­˜åœ¨ - è·³è¿‡")

if copied:
    print(f"âœ… å·²å¤åˆ¶ {len(copied)} ä¸ª notebook åˆ°æ‚¨çš„å·¥ä½œç©ºé—´:")
    for nb in copied:
        print(f"  ğŸ“ {nb}")
    print(f"\nğŸ“ ä½ç½®: {my_notebooks}")
else:
    print("â„¹ï¸  æ‰€æœ‰ notebook å·²å¤åˆ¶æˆ–æºæœªæ‰¾åˆ°")
    print("\nğŸ’¡ æç¤º: å¦‚éœ€è¦ï¼Œæ‚¨å¯ä»¥ä» /shared/notebooks/ æ‰‹åŠ¨å¤åˆ¶")
```
